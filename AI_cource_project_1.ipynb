```python
import pandas as pd
import numpy as np
reviews_df=pd.read_csv('moviereviews.tsv', sep='\t')
print("Preview of the dataset:\n", reviews_df.head())
```

```python
print("Missing values in each column:\n", reviews_df.isnull().sum())
```

```python
reviews_df.dropna(inplace=True)
print("Missing values in each column:\n", reviews_df.isnull().sum())
print("Label distribution after cleaning:\n", reviews_df['label'].value_counts())
```

```python
blank_indices = []

for i,lb,rv in reviews_df.itertuples():
    if type(rv)==str:
        if rv.isspace():
            blank_indices.append(i)
reviews_df.drop(blank_indices, inplace=True)
print("Label distribution after cleaning:\n", reviews_df['label'].value_counts())
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
reviews_df.head()
```

```python
sid.polarity_scores(reviews_df.loc[2]['review'])
```

```python
reviews_df.loc[2]['label']
```

```python
reviews_df['scores']=reviews_df['review'].apply(lambda review : sid.polarity_scores(review))
reviews_df.head()
```

```python
reviews_df['compound']=reviews_df['scores'].apply(lambda dict: dict['compound'])
reviews_df.head()
```

```python
reviews_df['sid_label']=reviews_df['compound'].apply(lambda compound: 'pos' if compound>=0 else 'neg')
reviews_df.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(reviews_df['label'],reviews_df['sid_label']))
```

```python
print(confusion_matrix(reviews_df['label'],reviews_df['sid_label']))
```

```python
print(classification_report(reviews_df['label'],reviews_df['sid_label']))
```
